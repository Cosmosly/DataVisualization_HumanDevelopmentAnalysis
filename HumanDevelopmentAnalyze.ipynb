## Create the heatmap of human_development

```python
import pandas as pd
import plotly.express as px

# parse GNI to numeric 64,992 is a string and cannot parse to 64992 because the comma
# convert the object to numeric
a = human_development['Gross National Income (GNI) per Capita']
comma = ","
a1 = []
for element in a :
  if comma in element:
    element = element.translate({ord(','): None})
    a1.append(element)
  else:
    a1.append(element)
human_development['Gross National Income (GNI) per Capita'] = a1
human_development['Gross National Income (GNI) per Capita'] = human_development['Gross National Income (GNI) per Capita'].apply(pd.to_numeric)
```

```python
human_development.head(1)
```

```python
human_development.info()
```

```python
# generate the correlation
HD_corr = human_development[['Life Expectancy at Birth', 'Expected Years of Education',
       'Mean Years of Education', 'Gross National Income (GNI) per Capita',
       'GNI per Capita Rank Minus HDI Rank']]
HD_corr = HD_corr.corr()
# plot the heatmap
px.imshow(HD_corr,text_auto=True, color_continuous_scale='RdBu_r', origin='lower', title ="Human Development Correlation",aspect = 'auto')
```

#### As you can see above, Life Expectancy at Birth & Mean Years of Education &Expected Years of Education these three have strong positive linear correlation.
#### GNI per Capita with thoes three have good posiive liner correlation.

# Question 1: What are the top 30 education contries? Showing their locations, means education years.

```python
import plotly.express as px
# import the gapminder dataset
df = px.data.gapminder()
df.columns
```

```python
df['iso_alpha']
```

### I want to merge two table and use the iso_alpha code

```python
# change the name from coountry to Country
df.columns = ['Country', 'continent','year','lifeExp', 'pop', 'gdpPercap','iso_alpha', 'iso_num']
# select the columns we want to merge
columns = ['Country', 'iso_alpha']
HM_merge = human_development.merge(df[columns], how='right', on='Country')
```

```python
fig = px.choropleth(HM_merge, locations="iso_alpha", color="Mean Years of Education", hover_name="Country", title = 'Average Education Years in the World')
fig.show()
```

```python
# drop the duplicates columns cause by merge
HM_merge_1 = HM_merge.drop_duplicates(keep='first')
# sort the education years and selct top 30
HM_merge_20 = HM_merge_1.sort_values(by='Mean Years of Education', ascending=False).iloc[:30,:]
```

```python
fig = px.choropleth(HM_merge_20, locations="iso_alpha", color="Mean Years of Education", hover_name="Country", title = 'Top 30 Average Education Years in the World')
fig.show()
```

### Using ipywidgets to create a change variable plot

```python
import ipywidgets
import plotly.express as px


# define the function
def global_plot(color = 'Mean Years of Education'):
  fig = px.choropleth(HM_merge, locations="iso_alpha", color= color, hover_name="Country")
  fig.show()
```

```python
# call the ipywideges.interact
ipywidgets.interact(global_plot, color = ['HDI Rank','Human Development Index (HDI)',
       'Life Expectancy at Birth', 'Expected Years of Education',
       'Mean Years of Education', 
       'GNI per Capita Rank Minus HDI Rank', 
       'pop', 'gdpPercap'])
```

```python
# List the one party contries
one_party_contries = ['China', 'Hong Kong, China (SAR)', 'Russia', 'Iran (Islamic Republic of)','Myanmar','Cuba','Viet Nam','Eritrea' ]
# 1. 'Vietnam' in the gapminder should Change to 'Viet Nam'
# 2. 'Hong Kong, China' --> 'Hong Kong, China (SAR)'
# 3. 'Iran' --> 'Iran (Islamic Republic of)'
# 4. not Laos, not north korea, not Sahrawi Arab Democratic Republic
# 5. add Russia iso_alpah = RU

# import px and gapminder dataset
import plotly.express as px
# import the gapminder dataset
df2 = px.data.gapminder()
# drop the duplicate countries
df2 = df2.drop_duplicates(subset=['country'])
# change the countries name
df2['country'] = df2['country'].replace({'Vietnam':'Viet Nam', 'Hong Kong, China':'Hong Kong, China (SAR)', 'Iran':'Iran (Islamic Republic of)'})
# add russia
russian = pd.Series({'country':'Russia','iso_alpha':'RU'})
df2 = pd.concat([df2,russian.to_frame().T], ignore_index=True)
# change the name from coountry to Country
df2.columns = ['Country', 'continent','year','lifeExp', 'pop', 'gdpPercap','iso_alpha', 'iso_num']
df2
```

### Merge the dataset
